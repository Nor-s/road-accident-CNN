```python
import cv2
import os

def make_dirs(dirs):
    if not os.path.exists(dirs):
        os.makedirs(dirs)
        
def center_crop(img, set_size):

    h, w = img.shape

    if set_size > min(h, w):
        return img

    crop_width = set_size
    crop_height = set_size

    mid_x, mid_y = w//2, h//2
    offset_x, offset_y = crop_width//2, crop_height//2
       
    crop_img = img[mid_y - offset_y:mid_y + offset_y, mid_x - offset_x:mid_x + offset_x]
    return crop_img

def process_crop(dir, size, save_path):
    make_dirs(save_path)
    file_list = os.listdir(dir)
    for filename in file_list:
        src_filename = dir + '\\' + filename
        img = cv2.imread(src_filename, cv2.IMREAD_GRAYSCALE)
        img = center_crop(img, size)
        # img = cv2.equalizeHist(img)
        cv2.imwrite(save_path + '//' + filename, img)

def remove_file(dir):
    file_list = os.listdir(dir)
    for filename in file_list:
        src_filename = dir + '\\' + filename
        img = cv2.imread(src_filename, cv2.IMREAD_GRAYSCALE)
        if img.sum() > 13873000:
            os.remove(src_filename)
            
        
size = 256
```
